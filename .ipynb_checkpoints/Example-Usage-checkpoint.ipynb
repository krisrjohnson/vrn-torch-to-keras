```python
from keras.models import load_model
import cv2
import visvis as vv
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = [40,40]

import custom_layers
```

# Example usage of vrn-unguided-keras.h5
Download the h5py file here:
https://drive.google.com/file/d/1oh8Zpe4wh00iXcm8ztRsi5ZL6GMkHdjj/view?usp=sharing

```python
custom_objects = {
    'Conv': custom_layers.Conv,
    'BatchNorm': custom_layers.BatchNorm,
    'UpSamplingBilinear': custom_layers.UpSamplingBilinear
}
model = load_model('vrn-unguided-keras.h5', custom_objects=custom_objects)
```

```python
model.summary()
```

```python
matplotlib.rcParams['figure.figsize'] = [20,20]
img2 = cv2.imread('dr_mark.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
print(img2.shape)
plt.imshow(img2.squeeze().transpose(0,1,2))
```

```python
img = cv2.imread('dr_mark.jpg')
img = img[800:2400, 424:2024] #1224
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
smallside = min(img.shape[:2])
longside = max(img.shape[:2])
startpt, endpt = longside//2 - smallside // 2, longside//2 + smallside//2
#square crop, centered (can do much more easily w/ torch.transforms)
print(smallside, longside, startpt, endpt)
if (img.shape[0] == smallside): img = img[0:smallside, startpt:endpt]
elif (img.shape[1] == smallside): img = img[startpt:endpt, 0:smallside]
else: print('broken')

img = cv2.resize(img, (192, 192))
img = img.transpose(2,0,1) #move from W*H*n_C to n_C*W*H
img = np.expand_dims(img, axis=0)
img.shape
```

```python
plt.imshow(img.squeeze().transpose(1,2,0))
```

```python
pred = model.predict(img)
```

```python
vol = pred[0] * 255
im = img[0]
im = np.swapaxes(im, 0, 1)
im = np.swapaxes(im, 1, 2)

volRGB = np.stack(((vol > 1) * im[:,:,0],
                   (vol > 1) * im[:,:,1],
                   (vol > 1) * im[:,:,2]), axis=3)
```

```python
for i, p in enumerate(volRGB[80:180:5]):
    plt.subplot(5, 4, i + 1)
    plt.imshow(p)
    
    
plt.show()
```

```python
vv.clf()

t = vv.imshow(im)
t.interpolate = True

v = vv.volshow(volRGB, renderStyle='iso')

l0 = vv.gca()
l0.light0.ambient = 0.9 # 0.2 is default for light 0
l0.light0.diffuse = 1.0 # 1.0 is default

a = vv.gca()
a.camera.fov = 0 # orthographic

vv.use().Run()
```
